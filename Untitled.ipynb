```python
# Feature engineering function
def feat_eng(df):
    # Add sequence length
    df['seq_length'] = df['sequence'].str.len()
    
    # Count single nucleotides
    df['A_cnt'] = df['sequence'].str.count('A')
    df['C_cnt'] = df['sequence'].str.count('C')
    df['U_cnt'] = df['sequence'].str.count('U')
    df['G_cnt'] = df['sequence'].str.count('G')
    
    # Count nucleotide pairs
    pairs = ['AC', 'AU', 'AG', 'CA', 'CU', 'CG',
             'UA', 'UC', 'UG', 'GA', 'GC', 'GU',
             'AA', 'CC', 'UU', 'GG']
    for pair in pairs:
        df[f'{pair}_cnt'] = df['sequence'].str.count(pair)
    
    # Extract the first and last characters of the sequence
    df['begin_seq'] = df['sequence'].str[0]
    df['end_seq'] = df['sequence'].str[-1]
    
    # Drop unnecessary columns
    cols_to_drop = ['sequence', 'temporal_cutoff', 'description', 'all_sequences']
    df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])
    return df

# Apply the feature engineering function
train_sequences = feat_eng(train_sequences)

# Create the target_id from ID
train_labels['target_id'] = train_labels['ID'].apply(lambda x: '_'.join(x.split('_')[:2]))

# Merge train_labels and train_sequences
train_data_clean = pd.merge(train_labels, train_sequences, on='target_id', how='left')

# Impute missing values for x_1, y_1, z_1 with group averages
for col in ['x_1', 'y_1', 'z_1']:
    train_data_clean[col] = train_data_clean.groupby(['target_id', 'resname'])[col].transform(
        lambda x: x.fillna(x.mean())
    )

# Drop remaining rows with NAs
train_data_clean = train_data_clean.dropna()

# Print the number of NAs left (should be 0)
print(f"There are {train_data_clean.isna().sum().sum()} NA's in the dataset!")
```

```python
# Initialize the H2O cluster
h2o.init()

# Import your cleaned dataset (replace with actual file path)
train_data_clean = h2o.import_file("train_data_clean")

# Function to train XGBoost model for a specific target
def train_xgboost_model(data, target, excluded_columns):
    # Remove unnecessary columns
    train_hframe = data.drop(excluded_columns)
    
    # Convert categorical variables to factors
    train_hframe["resname"] = train_hframe["resname"].asfactor()
    train_hframe["begin_seq"] = train_hframe["begin_seq"].asfactor()
    train_hframe["end_seq"] = train_hframe["end_seq"].asfactor()
    
    # Split data into training and validation sets
    splits = train_hframe.split_frame(ratios=[0.8], seed=1)
    train = splits[0]
    valid = splits[1]
    
    # Define features (exclude the target)
    features = [col for col in train_hframe.columns if col != target]
    
    # Train the XGBoost model
    model = H2OXGBoostEstimator(
        ntrees=500,
        max_depth=5,
        booster="gbtree",
        nfolds=5,
        seed=1,
        keep_cross_validation_predictions=True
    )
    model.train(x=features, y=target, training_frame=train, validation_frame=valid)
    
    # Print Mean Squared Error
    mse = model.mse(train=True, valid=True)
    print(f"MSE for {target}: {mse}")
    return model

# Train model for x_1
xgb_x = train_xgboost_model(train_data_clean, target="x_1", excluded_columns=["ID", "y_1", "z_1"])

# Train model for y_1
xgb_y = train_xgboost_model(train_data_clean, target="y_1", excluded_columns=["ID", "x_1", "z_1"])

# Train model for z_1
xgb_z = train_xgboost_model(train_data_clean, target="z_1", excluded_columns=["ID", "y_1", "x_1"])

# Shut down H2O cluster when done
h2o.shutdown(prompt=False)
```
